```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
# For Microsoft Edge Browser

import os
from selenium import webdriver
from selenium.webdriver.edge.service import Service

# provide path to the Edge driver
s=Service("C:\\Users\sandr\OneDrive\Desktop\Drivers\msedgedriver.exe")
driver = webdriver.Edge(service=s)

# Get the url
driver.get("https://www.nba.com/stats/draft/combine-anthro/")

# Get the page source or source code of the webpage
print(driver.page_source[:100])
```

```python
# Use BeautifulSoup and lxml extension to parse, save in variable nba_soup
nba_soup = BeautifulSoup(driver.page_source, 'lxml')
```

```python
# Players
list_players = []
for item in nba_soup.find_all('td', class_='first'):
    list_players.append(item.text)

list_players
```

```python
# Number of players on one table
len(list_players)
```

```python
# Locate the table on the page
table = nba_soup.find('table')
table
```

```python
# Obtain the header for each column on the table
# Clean up headers by removing unnecessary characters

headers = []
for i in table.find_all('th'):
    title = i.text.replace('\n', '').replace('\xa0', ' ').strip()
    headers.append(title)
    

headers # list of column names
```

```python
# Create a dataframe
nba_stat = pd.DataFrame(columns = headers) # use header list as column names for new dataframe
```

```python
# Create a for loop to fill nba_stat
for i in table.find_all('tr')[1:]:
    row_data = i.find_all('td')
    row =[j.text for j in row_data]
    length = len(nba_stat)
    nba_stat.loc[length] = row
```

## Clean up the data

```python
# Make a copy of the dataframe, jic
nba_stat1 = nba_stat.copy()
```

```python
# Rename some column names to include units of measure
nba_stat1 = nba_stat1.rename(columns = {"HEIGHT W/O SHOES":"HEIGHT W/O SHOES (inches)",
                                        "HEIGHT W/ SHOES": "HEIGHT W/ SHOES (inches)",
                                       "STANDING REACH": "STANDING REACH (inches)",
                                        "WEIGHT (LBS)": "WEIGHT (lbs)",
                                       "WINGSPAN": "WINGSPAN (inches)"})
nba_stat1.head(10)
```

```python
# Remove column POS, not significant
nba_stat1 = nba_stat1.drop(['POS'], axis=1)
nba_stat1
```

```python
# Remove '%' from each row in "BODY FAT %" column
nba_stat1['BODY FAT %'] = [item.replace('%','') for item in nba_stat1['BODY FAT %']]
nba_stat1.head()  
```

```python
# Find null counts
null_counts = nba_stat1.isna().sum()
null_counts
```

```python
# Drop all null values
nba_stat1 = nba_stat1.dropna()

# Check dimensions after dropping null values
nba_stat1.shape
```

```python
# Make a function to standardize height and wingspan into inches
def convert_to_inches(measure):
        feet_to_inches = float(measure.split("'")[0])*12
        inches = float(measure.split("'")[1].replace("'","").replace('"', ''))
        total_inches = feet_to_inches + inches
        return total_inches


#for testing:
# measure1 = "6' 5.75''"
# measure2 = "6' 4.50'"
# convert_to_inches(measure2)
# convert_to_inches(measure1)
```

```python
# Apply conversion function to each column that needs conversion
nba_stat1['HEIGHT W/O SHOES (inches)'] = nba_stat1['HEIGHT W/O SHOES (inches)'].apply(convert_to_inches)
nba_stat1['HEIGHT W/ SHOES (inches)'] = nba_stat1['HEIGHT W/ SHOES (inches)'].apply(convert_to_inches)
nba_stat1['STANDING REACH (inches)'] = nba_stat1['STANDING REACH (inches)'].apply(convert_to_inches)
nba_stat1['WINGSPAN (inches)'] = nba_stat1['WINGSPAN (inches)'].apply(convert_to_inches)
nba_stat1.head()
```
