```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Request the 
nba_url = "https://www.nba.com/stats/draft/combine-anthro/"

# retrieve text file code (raw data) from url using request
page = requests.get(nba_url)
```

```python
# Check the status code
page.status_code # 200 means success
```

```python
# use .text to transform html into python string, save to variable 'nba_stat'
nba_stat = page.text
```

```python
# Use BeautifulSoup and lxml extension to parse
nba_soup = BeautifulSoup(nba_stat, 'lxml')

nba_soup
```
